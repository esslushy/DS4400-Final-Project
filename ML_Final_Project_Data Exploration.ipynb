```python
def heatmap(dataframe, start_col, end_col, color):
    df_features = dataframe.iloc[:, start_col:end_col]
    
    # Find the correlation coefficient between each trait as a new dataframe (absolute value applied so all coefficients are between 0 and 1)
    df_corr = df_features.corr().abs()
    
    # Set sizes of figure
    fig_dims = (10, 7)
    fig, ax = plt.subplots(figsize=fig_dims)
    
    # Create a blue heatmap of strengh of correlation
    graph = sns.heatmap(df_corr, vmin=0, vmax=1, linewidths=0.5, cmap=color, ax=ax)
    graph.set(title="Correlation between Features")
```

```python
heatmap(df, 0, 10, 'Blues')
```

```python
def scatterplot(df, x_col, y_col, title, x_label, y_label, hover_col):
    fig = px.scatter(df, x=x_col, y=y_col, title=title, hover_name=hover_col)
    
    # modify axis label properties
    fig.update_xaxes(title=x_label,
                     title_font={"size":16, "color":"gray"}, 
                     tickfont = {"size":14, "color":"gray"})
    fig.update_yaxes(title=y_label,
                     title_font={"size":16, "color":"gray"}, 
                                tickfont = {"size":14, "color":"gray"})

    fig.show()
```
